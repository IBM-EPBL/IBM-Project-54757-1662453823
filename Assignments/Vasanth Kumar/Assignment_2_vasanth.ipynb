# Handling the missing data and outliers

```python
sns.boxplot(ds.CreditScore,color='pink')
```

```python
q1=ds.CreditScore.quantile(0.25)  
q3=ds.CreditScore.quantile(0.75)

IQR=q3-q1

upper_limit= q3 + 1.5*IQR
lower_limit= q1 - 1.5*IQR
```

```python
print("Upper limit :",upper_limit)
print("Lower limit :",lower_limit)
```

```python
ds.median()
```

```python
ds['CreditScore']= np.where(ds['CreditScore']<lower_limit,6.520000e+02,ds['CreditScore'])
```

```python
sns.boxplot(ds.CreditScore,color='pink')
```
